# Data handling & cleaning

```python
df.drop(columns=["Unnamed: 0","order_id","customer_id","product_id","zipcode"],inplace=True) # Removing colunmns for better readability
```

# EDA & Visualization

## Payment method classification

```python
df["payment_method"].value_counts().plot(kind="pie",autopct="%1.1f%%",figsize=(8,8),pctdistance=0.85,fontsize=10,startangle=160,color="colors",title="Payment Method Classification")
```

```python
df["product_category"].value_counts().plot(kind="pie",autopct="%1.1f%%",figsize=(8,8),pctdistance=0.85,fontsize=10,startangle=160,color="colors",title="Product Categrory Classification")
```

## Histogram Distrubution of Product category,payment method and delivery status

```python
obj=df.drop(columns=["city","state"]).select_dtypes(include="object")
for col in obj:
    plt.figure(figsize=(12,6))
    sns.histplot(data=df,x=col,kde=True,hue="product_category",edgecolor="black")
    plt.title(f"The Distribution of {col}")
    plt.show()
```

## Total product price per category

```python
classifi_pr=df.groupby("product_category")["product_price"].sum().sort_values(ascending=False).reset_index()
classifi_pro=classifi_pr.style.background_gradient(cmap="Reds")
classifi_pro
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=classifi_pr,x="product_category",y="product_price",hue="product_category",edgecolor="black")
plt.tight_layout()
plt.show()
```

## Which product category has high demand

```python
classifi_dem=df.groupby("product_category")["quantity"].sum().sort_values(ascending=False).reset_index()
classifi_demand=classifi_dem.style.background_gradient(cmap="Oranges")
classifi_demand
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=classifi_dem,x="product_category",y="quantity",hue="product_category",edgecolor="black",palette="viridis")
plt.tight_layout()
plt.legend()
plt.show()
```

## Which city and state prefers which payment method
#### *This way we can understand that which payment method customer's wanna use and we can offer discount on products basis on that*

```python
classifi_st = df.groupby(["state", "city", "payment_method"]).size().reset_index(name="count")
classifi_st
# Pivot table: cities vs payment methods
city_payment = classifi_st.pivot_table(
    index="city", columns="payment_method", values="count", fill_value=0
)

# Optional: show top 20 cities only
top_cities = city_payment.sum(axis=1).nlargest(50).index
city_payment_top = city_payment.loc[top_cities]

# Plot
plt.figure(figsize=(15,12))
sns.heatmap(city_payment_top, annot=True, fmt='g', cmap='YlGnBu')
plt.title("Top 50 Cities vs Payment Method Usage")
plt.xlabel("Payment Method")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

## Average Discount PerProduct Category

```python
classifi_dis=df.groupby("product_category")["discount_applied"].mean().sort_values(ascending=False).reset_index()
classifi_dis
plt.figure(figsize=(12,6))
sns.barplot(data=classifi_dis,x="product_category",y="discount_applied",hue="product_category",edgecolor="black")
plt.title("Average Discount PerProduct Category",size=13)
plt.show()
```

## Average discount value and total order value per category

```python
classifi_ca=df.groupby("product_category").agg({
    "discount_applied":"mean",
    "order_value":"sum"
}).reset_index()
classifi_ca
fig,axs=plt.subplots(1,2,figsize=(18,6))
sns.barplot(data=classifi_ca,x="product_category",y="discount_applied",ax=axs[0])
axs[0].set_title("Average Discount Value Per Category",size=13)
sns.barplot(data=classifi_ca,x="product_category",y="order_value",ax=axs[1])
axs[1].set_title("Total Order Value Per Category",size=13)
plt.show()
```

## Return Request Count Per Product Category

```python
classifi_ret=df.groupby(["product_category","return_requested"]).size().reset_index(name="count")
plt.figure(figsize=(12,6))
sns.barplot(data=classifi_ret,x="return_requested",y="count",hue="product_category",edgecolor="black")
plt.title("Return Request Count Per Product Category",size=13)
plt.show()
```

## Orders Over Time

```python
orders_by_date = df["order_date"].value_counts().sort_index()

# Plot
plt.figure(figsize=(20,6))
orders_by_date.plot(kind="line", marker="o", linestyle="-",color="red")
plt.title("Orders Over Time")
plt.xlabel("Order Date")
plt.ylabel("Number of Orders")
plt.grid()
plt.show()
```
