# EDA And Visualization

```python
plt.figure(figsize=(12,6))
sns.histplot(df["age"], bins=30, kde=True, color="teal")
plt.title("Age Distribution of Users")
plt.show()

city_counts = df["city"].value_counts().head(10)
sns.barplot(x=city_counts.index, y=city_counts.values, palette="viridis")
plt.title("Top 10 Cities by Orders")
plt.xticks(rotation=45)
plt.show()
```

```python
cuisine_counts = df["cuisine"].value_counts().head(10)
sns.barplot(x=cuisine_counts.index, y=cuisine_counts.values, palette="magma")
plt.title("Top 10 Cuisines Ordered")
plt.xticks(rotation=45)
plt.show()

meal_counts = df["meal_type"].value_counts()
meal_counts.plot(kind="pie", autopct="%1.1f%%", startangle=90, figsize=(8,8))
plt.title("Meal Type Distribution")
plt.ylabel("")
plt.show()
```

```python
rest_type_sales = df.groupby("restaurant_type")["order_value"].sum().sort_values(ascending=False)
sns.barplot(x=rest_type_sales.index, y=rest_type_sales.values, palette="coolwarm")
plt.title("Total Sales by Restaurant Type")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df["order_value"], bins=50, kde=True, color="green")
plt.title("Distribution of Order Value")
plt.show()

sns.boxplot(x="discount_applied", y="order_value", data=df, palette="Set2")
plt.title("Order Value vs Discount Applied")
plt.show()
```

```python
sns.scatterplot(x="delivery_fee", y="order_value", data=df, alpha=0.5)
plt.title("Delivery Fee vs Order Value")
plt.show()

sns.histplot(df["time_taken_to_order"], bins=30, kde=True, color="orange")
plt.title("Distribution of Time Taken to Place Order")
plt.show()
```

```python
sns.countplot(x="rating_given", data=df, palette="Blues")
plt.title("Distribution of Ratings")
plt.show()

repeat_counts = df["is_repeat_order"].value_counts()
repeat_counts.plot(kind="pie", autopct="%1.1f%%", startangle=90, figsize=(8,8))
plt.title("Repeat vs First Orders")
plt.ylabel("")
plt.show()
```

```python
sns.countplot(x="mood", data=df, palette="pastel")
plt.title("Orders by Mood")
plt.show()

sns.countplot(x="hunger_level", data=df, palette="rocket")
plt.title("Orders by Hunger Level")
plt.show()

sns.countplot(x="company", data=df, palette="muted")
plt.title("Orders by Company (Alone/Group/Family)")
plt.show()

sns.countplot(x="rainy_weather", data=df, palette="cool")
plt.title("Orders in Rainy vs Non-Rainy Weather")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[["order_value","delivery_fee","time_taken_to_order","rating_given","age"]].corr(),
            annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Numeric Features")
plt.show()
```
